## Alpha-beta Pruning Algorithm

Alpha is the best value that the maximizer currently can guarantee at that level or above.

Beta is the best value that the minimizer currently can guarantee at that level or below.

```python
#Initially called for root and maximizer
def minimax(depth, nodeIndex, maximizingPlayer, alpha, beta): 
    
    # Stopping condition (leaf node is reached) 
    if depth == NUMLEVELS: 
        return VALUES[nodeIndex] 

    if maximizingPlayer: 
        best = MIN
        # Recursion for left and right children 
        for i in range(0, 2): 
            val = minimax(depth + 1, nodeIndex * 2 + i, False, alpha, beta) 
            best = max(best, val) 
            alpha = max(alpha, best) 

            # Alpha Beta Pruning 
            if beta <= alpha: 
                break
        return best 

    else:
        best = MAX
        # Recursion for left and right children 
        for i in range(0, 2): 
            val = minimax(depth + 1, nodeIndex * 2 + i, True, alpha, beta) 
            best = min(best, val) 
            beta = min(beta, best) 

            # Alpha Beta Pruning 
            if beta <= alpha: 
                break
        return best 
```

```python
# Initial values of Alpha and Beta 
MAX, MIN = 1000, -1000
NUMLEVELS = 2
#NUMLEVELS = 4

#VALUES = [3, -14, 11, 0, -14, -9, -3, -3, -11, 7, -16, -11, 15, 3, -17, -5] 
#VALUES = [9, -19, -18, 19, -18, -7, -8, 4, 16] 
VALUES = [2, 8, 9, -10] 

print("The optimal value is :", minimax(0, 0, False, MIN, MAX)) 
```
